```python
#Import necessary libraries:
import networkx as nx
from networkx.algorithms import community
import os
#Importing the data
import pandas as pd
import csv
#Graphing:
import matplotlib.pyplot as plt
import pylab as pl #For saving plots as images
import numpy as np
#To be used for dictionary lists:
import operator
from operator import itemgetter
```

Using Pandas to import data:

```python
df = pd.read_csv('DelawareSocialNetwork.csv')
#print(df)
```

Parse the edgelist from the CSV file so NetworkX will read it as node connections:

```python
Data = open('DelawareSocialNetwork.csv', "r")
Graphtype = nx.Graph()
W = nx.parse_edgelist(Data, comments='t', delimiter=',', create_using=Graphtype,
                      nodetype=int, data=(('weight', float),))
nx.draw(W)
plt.show()
print(nx.info(W)) 
```

Kamada Kawai - One of the best visualizations for larger social networks:

```python
plt.figure(figsize=(30,24))
nx.draw_kamada_kawai(W, with_labels=True, node_size=480, alpha=.9, weight=True, font_size=14, node_color='yellow')
plt.show()
print(nx.info(W))
```

In total there are 232 dolphins that were photo identified in the study, but the social network is only composed of 222. Some of the dolphins in certain sightings were dolphins that did not interact with any other dolphins in the network except each other. In total there were 10 individuals.<br>
<br>List the nodes that make up the social network:

```python
W.nodes()
```

Explore the top ten various degrees and ranks in the social network including the overall clustering coefficient and density:

```python
print("Average Clustering Coefficient:",nx.average_clustering(W))
print("Density:",nx.density(W))
```

```python
#Top 10 by Page Rank:
PR = nx.pagerank(W)
Dict_PR = sorted(PR.items(), key=itemgetter(1), reverse=True) 
print("Top 10 by Page Rank:")
print("")
for i in Dict_PR[:10]:
    print(i)
```

```python
#Top 10 by Average Degree Connectivity:
Degree=nx.average_degree_connectivity(W)
Dict_Deg = sorted(Degree.items(), key=itemgetter(1), reverse=True) 
print("Average Degree Connectivity Top 10:")
print("")
for i in Dict_Deg[:11]:
    print(i)
```

```python
#Top 10 by Degree Centality:
Degree=nx.degree_centrality(W)
Dict_Deg = sorted(Degree.items(), key=itemgetter(1), reverse=True) 
print("Degree Centrality Top 10:")
print("")
for i in Dict_Deg[:10]:
    print(i)
plt.show()
```

```python
#Top 10 by Betweenness Centrality:
BC=nx.betweenness_centrality(W)
Dict_BC = sorted(BC.items(), key=itemgetter(1), reverse=True) 
print("Top 10 Betweeness Centrality:")
print("")
for i in Dict_BC[:10]: #items //appending vs. printing
    print(i) 
```

```python
#Top 10 by Closeness Centrality:
BC=nx.closeness_centrality(W)
Dict_BC = sorted(BC.items(), key=itemgetter(1), reverse=True) 
print("Top 10 Closeness Centrality:")
print("")
for i in Dict_BC[:10]: #items //appending vs. printing
    print(i)
```

```python
#Top 10 by Eigenvector Centrality:
BC=nx.eigenvector_centrality(W)
Dict_BC = sorted(BC.items(), key=itemgetter(1), reverse=True) 
print("Top 10 Eigenvector Centrality:")
print("")
for i in Dict_BC[:10]: #items //appending vs. printing
    print(i) 
```

Graph the social network including the individuals that do not interact with the main social network, and graph them by betweenness centrality degree with a specific color scale:
<br>Reference resource: https://www.datacamp.com/community/tutorials/social-network-analysis-python (Below:)

```python
pos = nx.spring_layout(W)
betCent = nx.betweenness_centrality(W, normalized=True, endpoints=True)
node_color = [20000 * W.degree(v) for v in W]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(30,30))
nx.draw_networkx(W, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')

print("Nodes with highest betweenness centrality:",sorted(betCent, key=betCent.get, reverse=True)[:10])
```
